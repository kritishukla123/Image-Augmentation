# Image augmentation

```python
import imageio
import imgaug as ia
import imgaug.augmenters as iaa
import matplotlib.pyplot as plt
```

```python
image=imageio.imread(r'C:\Users\DELL\Pictures\Saved Pictures\download.jpg')
ia.imshow(image)
```

## Rotate Image

```python
rotate=iaa.Affine(rotate=(-50,10))
rotated_image=rotate.augment_image(image)
ia.imshow(rotated_image)
```

## Shearing of image(move one side of the image)

```python
shear=iaa.Affine(shear=(20))
shear_image=shear.augment_image(image)
ia.imshow(shear_image)
```

## Adding noise to the image 

```python
gausian_noise=iaa.AdditiveGaussianNoise(0,200)
gausian_image=gausian_noise.augment_image(image)
ia.imshow(gausian_image)
```

## Scaling of Image (zoom in/out)

```python
scale=iaa.Affine(scale=0.9)
scale_image=scale.augment_image(image)
ia.imshow(scale_image)
```

```python
invert=iaa.Invert(0.8,per_channel=0.2)
invert_image=invert.augment_image(image)
ia.imshow(invert_image)
```
